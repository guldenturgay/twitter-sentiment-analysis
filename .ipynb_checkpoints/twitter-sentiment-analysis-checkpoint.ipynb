## Reading and cleaning the data

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
```

```python
dataset = pd.read_csv("data/Tweets.csv", sep=",")
dataset.head()
```

```python
dataset = dataset[['airline_sentiment','text']]
dataset.head()
```

```python
import re
import string

# Stopwords in English language
stopword = nltk.corpus.stopwords.words('english')

# Lemmatizer
wn = nltk.WordNetLemmatizer()

# Function to clean and tokenize the data
def clean_tokenize_text(text):
    # Remove mentions starting with @
    text = re.sub('@\w+\s', '', text)
    
    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    
    # Tokenize the text
    tokens = re.split('\W+', text)
    
    # Remove stopwords
    text = [word for word in tokens if word not in stopword]
    
    # Lemmatize the words
    text = [wn.lemmatize(word) for word in text]
    
    # Return text
    return text


# Apply the function to text column
dataset['text'] = dataset['text'].apply(lambda x: clean_tokenize_text(x.lower()))

# Rename the columns
dataset.columns=['label','text']

# Display first five rows
dataset.head()
```
